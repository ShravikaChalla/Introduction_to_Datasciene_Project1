# Reading the dataset

```python
import pandas as pd
# Read the CSV file into a DataFrame
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

#### Dataset Name: Electric Vehicle Population Data
Source: Washington State Department of Licensing

Columns:
1. VIN (1-10) : Unique Vehicle Identification Number (String)
2. County : County where the vehicle is registered (String)
3. City : City where the vehicle is registered (String)
4. State : State where the vehicle is registered (String)
5. Postal Code : Postal code of the registration location (Numeric)
6. Model Year : Year of the vehicle's model (Numeric)
7. Make : Manufacturer or brand of the vehicle (String)
8. Model : Model name of the vehicle (String)
9. Electric Vehicle Type : Type of electric vehicle (String)
10. Clean Alternative Fuel Vehicle (CAFV) Eligibility : Eligibility for Clean Alternative Fuel Vehicle (String)
11. Electric Range : Range of the vehicle on electric power (Numeric)
12. Base MSRP : Manufacturer's Suggested Retail Price (Numeric)
13. Legislative District : Legislative district associated with the vehicle (Numeric)
14. DOL Vehicle ID : Unique ID assigned by the Department of Licensing (Numeric)
15. Vehicle Location : Location information of the vehicle (String)
16. Electric Utility : Electric utility associated with the vehicle (String)
17. 2020 Census Tract : Census tract of the registration location for 2020 (Numeric)

# Data Cleaning

Data cleaning, which involves finding and fixing discrepancies, errors, and missing or inconsistent entries within a dataset, is a crucial step in the data analysis process. This process makes sure that the data used for analysis is precise, trustworthy, and appropriate for producing valuable insights. Handling missing values, dealing with duplicates, confirming and adjusting data types, managing outliers, standardizing units and formats, and fixing any inconsistencies are all tasks that fall under the category of data cleaning.


```python
#Handling Missing Values
df['Electric Range'].fillna(df['Electric Range'].mean(), inplace=True)
#Removing Duplicates
df.drop_duplicates(inplace=True)
#Correcting Inconsistencies (if any)
df['Make'] = df['Make'].str.lower()
df.info()
```

## Electric range of vehicles

```python
# Assuming 'df' is your DataFrame and 'Electric Range' is the column you want to analyze
plt.hist(df['Electric Range'], bins=30)
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.title('Histogram of Electric Range')
plt.show()
```

#### Electric range is an important consideration for consumers when choosing an electric vehicle, as it determines the vehicle's suitability for their typical driving needs and habits. Manufacturers often strive to improve the electric range of their vehicles to make them more competitive in the market.

```python
# Scatter plot for Model Year vs. Electric Range
plt.figure(figsize=(10, 6))
plt.scatter(df['Model Year'], df['Electric Range'], alpha=0.5, color='blue')
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.title('Model Year vs. Electric Range')
plt.grid(True)

# Show the plot
plt.show()
```

#### The scatter plot shows a varied distribution of electric vehicles across different model years, indicating that electric range has evolved over time.

# Calculate and analyze basic statistics like mean, median, mode.

```python
# Example: Mean and Median of Electric Range
mean_range = df['Electric Range'].mean()
median_range = df['Electric Range'].median()
print(f"Mean Electric Range: {mean_range}")
print(f"Median Electric Range: {median_range}")
```

```python
# Assuming 'df' is your DataFrame
make_counts = df['Make'].value_counts()

# Set a threshold for the number of makes to include
top_n = 10

# Group all makes below the threshold as 'Other'
make_counts['Other'] = make_counts[top_n:].sum()
make_counts = make_counts[:top_n]

plt.figure(figsize=(8, 8))
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Top {top_n} Makes of Electric Vehicles')
plt.show()
```

```python
# Assuming 'df' is your DataFrame
model_counts = df['Model'].value_counts().head(10)  # Selecting the top 10 models

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(model_counts, labels=model_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(model_counts))))
plt.title('Top 10 Models Distribution')
plt.show()
```

# Relationships between  electric range vary with make and model

### This involves finding out if there's a pattern between the brand (make) of a vehicle and how far it can go on a full electric charge

```python
import seaborn as sns

# Set the style of the plot
sns.set_style('whitegrid')

# Create a figure and axis with specified size
plt.figure(figsize=(12, 8))

# Create the box plot
sns.boxplot(x='Make', y='Electric Range', data=df, palette='pastel')

# Set labels and title
plt.xlabel('Make', fontsize=12)
plt.ylabel('Electric Range', fontsize=12)
plt.title('Electric Range by Make', fontsize=15)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

The box plot helps us see how electric ranges are spread out among different car makers. It gives a clear visual of the range distribution and identifies any unusual data points.






```python
# Define the list of cities you want to include
selected_cities = ['Los Angeles', 'San Francisco', 'New York City', 'Seattle', 'Austin']

# Filter data for the selected cities
filtered_df = df[df['City'].isin(selected_cities)]

# Group the data by City and State, and count the occurrences
city_state_counts = filtered_df.groupby(['City', 'State']).size()

# Create a bar chart
plt.figure(figsize=(12, 6))
city_state_counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('City, State')
plt.ylabel('Number of Vehicles')
plt.title('Vehicle Distribution by City and State')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Assuming 'df' is your DataFrame with columns 'Base MSRP' and 'Electric Range'
plt.figure(figsize=(10, 6))
plt.scatter(df['Base MSRP'], df['Electric Range'], alpha=0.5)
plt.xlabel('Base MSRP')
plt.ylabel('Electric Range')
plt.title('Scatter Plot: Base MSRP vs. Electric Range')
plt.show()
```

## Vehicle Distribution by City and State

```python
# Assuming 'df' is your DataFrame with columns 'City' and 'State'
# Assuming 'df' is your DataFrame with columns 'City' and 'State'
city_state_counts = df.groupby(['City', 'State']).size().reset_index(name='Counts')

# Get the top 10 city-state combinations
top_10_city_state = city_state_counts.nlargest(20, 'Counts')

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_city_state['City'] + ', ' + top_10_city_state['State'], top_10_city_state['Counts'], color='skyblue')
plt.xlabel('City, State')
plt.ylabel('Number of Vehicles')
plt.title('Top 10 Vehicle Distribution by City and State')
plt.xticks(rotation=45)
plt.show()
```
